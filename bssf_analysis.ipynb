```python
from tsp_solve_backtracking import backtracking, backtracking_bssf
from utils import generate_network, Timer
```

```python
size = 10
seed = 200
network = generate_network(size, seed)[1]
backtracking_solution = backtracking(network, Timer())[-1]
bssf_solution = backtracking_bssf(network, Timer())[-1]
backtracking_score = backtracking_solution.score
backtracking_time = backtracking_solution.time
bssf_score = bssf_solution.score
bssf_time = bssf_solution.time

print(f"Backtracking solution {backtracking_score} in {backtracking_time:.6f} seconds")
print(f"BSSF solution {bssf_score} in {bssf_time:.6f} seconds")
```
