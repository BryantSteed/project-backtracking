```python
from tsp_solve_backtracking import greedy_tour
from performance import run_analysis
from utils import generate_network, Timer
```

```python
def metric_func(size: int) -> float:
    network = generate_network(size, seed=42)[1]
    timer = Timer()
    solutions = greedy_tour(network, timer)
    if timer.time_out():
        print(f"Timeout occurred for size {size}")
    return solutions[-1].score  # Return the cost of the last solution
```

```python
sizes = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

run_analysis(sizes, metric_func)
```
