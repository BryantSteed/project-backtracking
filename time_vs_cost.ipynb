```python
from utils import generate_network, Timer
from tsp_plot import plot_solutions
from tsp_solve_backtracking import greedy_tour, backtracking, backtracking_bssf, random_tour
```

```python
network = generate_network(15, seed=34)[1]
time_limit = 60
greedy_solutions = greedy_tour(network, Timer(time_limit))
backtracking_solutions = backtracking(network, Timer(time_limit))
bssf_solutions = backtracking_bssf(network, Timer(time_limit))
random_solutions = random_tour(network, Timer(time_limit))
solutions = {
    "Greedy": greedy_solutions,
    "Backtracking": backtracking_solutions,
    "BSSF": bssf_solutions,
    "Random": random_solutions
}
```

```python
plot_solutions(solutions, ax=None)
print(solutions["BSSF"])
```
