```python
from performance import run_analysis
from tsp_solve_backtracking import backtracking, backtracking_bssf
from utils import Timer, generate_network
import matplotlib.pyplot as plt
```

```python
sizes = [5, 10, 15]
back_runtimes = []
greedy_runtimes = []
for size in sizes:
    network = generate_network(size, seed=34)[1]
    back_time = backtracking(network, Timer())[-1].time
    greedy_time = backtracking_bssf(network, Timer())[-1].time
    back_runtimes.append(back_time)
    greedy_runtimes.append(greedy_time)
```

```python
plt.clf()
plt.plot(sizes, back_runtimes, label='Backtracking')
plt.plot(sizes, greedy_runtimes, label='BSSF Backtracking')
plt.xlabel('Network Size')
plt.ylabel('Runtime (seconds)')
plt.title('Backtracking vs BSSF Backtracking Runtime Comparison')
plt.legend()
plt.show()
```
