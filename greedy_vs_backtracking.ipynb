```python
import matplotlib.pyplot as plt
```

```python
backtracking_sizes = [5, 10]
backtracking_runtimes = [0.22, 1556.17]
greedy_sizes = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
greedy_runtimes = [0.26, 0.52, 1.25, 2.32, 4.42, 4.87, 10.86, 13.76, 14.33, 19.37]
```

```python
plt.clf()
plt.plot(backtracking_sizes, backtracking_runtimes, 'o-', label='Backtracking', color='blue')
plt.plot(greedy_sizes, greedy_runtimes, 'o-', label='Greedy', color='orange')
plt.xlabel('Input Size (n)')
plt.ylabel('Runtime (log(ms))')
plt.title('Greedy vs Backtracking Runtimes')
plt.yscale('log')
plt.legend()
plt.show()
```
