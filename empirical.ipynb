```python
from tsp_solve_backtracking import greedy_tour, backtracking
from performance import run_analysis
from utils import generate_network, Timer
```

```python
def metric_func(size: int) -> bool:
    network = generate_network(size, seed=42)[1]
    timer = Timer()
    solutions = backtracking(network, timer)
    if timer.time_out():
        print(f"Timeout occurred for size {size}")
        return False
    return True  # Return the cost of the last solution
```

```python
sizes = [5, 10, 15]

run_analysis(sizes, metric_func)
```
